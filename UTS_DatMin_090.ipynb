```python
import pandas as pd

train = pd.read_csv('data_training.csv')
print("Shape training:", train.shape)
print("Kolom training:", train.columns.tolist())
print("\n3 data pertama:")
print(train.head(3))
```

```python
test = pd.read_csv('data_testing.csv')
print("Shape testing:", test.shape)
print("Kolom testing:", test.columns.tolist())
print("\n3 data pertama:")
print(test.head(3))
```

```python
# Pisahkan fitur dan target dari data training
X_train = train.drop(['quality', 'Id'], axis=1)
y_train = train['quality']

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("\nContoh target (y_train):")
print(y_train.head(5))
```

```python
# Pisahkan fitur dan ID dari data testing
X_test = test.drop('Id', axis=1)
test_ids = test['Id']

print("X_test shape:", X_test.shape)
print("Jumlah ID testing:", len(test_ids))
print("\nContoh ID:")
print(test_ids.head(5))
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Normalisasi selesai!")
print("Data training setelah scaling:")
print("Mean:", X_train_scaled.mean().round(2))
print("Std:", X_train_scaled.std().round(2))
print("\nShape X_train_scaled:", X_train_scaled.shape)
print("Shape X_test_scaled:", X_test_scaled.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier

# Buat model
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Latih model
model.fit(X_train_scaled, y_train)

print("Model selesai dilatih!")
print("Jenis model:", type(model).__name__)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Prediksi ke data training
y_pred_train = model.predict(X_train_scaled)

# Hitung akurasi
akurasi = accuracy_score(y_train, y_pred_train)
print(f"Akurasi model pada data training: {akurasi:.4f} ({akurasi*100:.2f}%)")

print("\nLaporan Klasifikasi:")
print(classification_report(y_train, y_pred_train))
```

```python
# Prediksi ke data testing
prediksi = model.predict(X_test_scaled)

print("Prediksi selesai!")
print("Jumlah prediksi:", len(prediksi))
print("\n10 prediksi pertama:")
print(prediksi[:10])
print("\nDistribusi hasil prediksi:")
print(pd.Series(prediksi).value_counts().sort_index())
```

```python
# Buat dataframe submission
submission = pd.DataFrame({
    'Id': test_ids,
    'Quality': prediksi
})

print("Preview 10 baris pertama:")
print(submission.head(10))
```

```python
# Simpan ke CSV
submission.to_csv('hasilprediksi_090.csv', index=False)
print("File 'hasilprediksi_090.csv' telah dibuat!")
```
